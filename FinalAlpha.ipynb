```python
# Reading srt file and make pandas DataFrame.
from soustitle import Subtitle
import pandas as pd

srt = Subtitle('speech.srt')
res = srt.open()

s = []
e = []
t = []
st = []
et = []
for i in range(len(res)):
    s.append(res[i]['start_time'])
    e.append(res[i]['end_time'])
    t.append(res[i]['subtitle_text'])

for i in range(len(s)):
    l=s[i].split(':')
    l.pop(0)
    l.pop(2)
    j= "00"+":"+l[0]+':'+l[1]
    st.append(j)

    m=e[i].split(':')
    m.pop(0)
    m.pop(2)
    m= "00"+":"+m[0]+':'+m[1]
    et.append(m)

df = pd.DataFrame()
df['start_time'] = st
df['end_time'] = et
df['speech_text'] = t
df['start_time']=pd.to_datetime(df['start_time'], format="%H:%M:%S")
df['end_time']=pd.to_datetime(df['end_time'], format="%H:%M:%S")
df['duration'] = df['end_time']-df['start_time']  # Duration of each sentence
df['duration']=df['duration'].apply(lambda x:str(x).split(':')[2]).apply(lambda x: int(x)) # taking only seconds in int form
```
